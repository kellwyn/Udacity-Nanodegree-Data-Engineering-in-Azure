```
dfbronzetosilver = spark.read.format("delta").load("/delta/bronze_matches")
```

```
from pyspark.sql.functions import sha2, concat_ws
dfbronzetosilver = dfbronzetosilver.dropDuplicates(dfbronzetosilver.columns)
dfbronzetosilver = dfbronzetosilver.withColumn("hash", sha2(concat_ws("||", *dfbronzetosilver.columns), 256))

dfbronzetosilver.write.format("delta").mode("overwrite").saveAsTable("silver_matches")
```

```
dfsilvertogold = spark.table("silver_matches")
display(dfsilvertogold)

dfsilvertogold = dfsilvertogold.groupby("season","winner").count()
dfsilvertogold.write.format("delta").mode("overwrite").saveAsTable("gold_seasonwinner")
display(dfsilvertogold)
```
